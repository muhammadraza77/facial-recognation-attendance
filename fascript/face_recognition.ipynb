```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
vr='C:\\Users\\th\\Anaconda3\\envs\\env1\\Library\\etc\\haarcascades\\'
face_cascade = cv.CascadeClassifier(vr+'haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier(vr+'haarcascade_eye.xml')
img = cv.imread('temp.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv.imshow('img',gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
faces = face_cascade.detectMultiScale(gray , 1.3 , 5)
print(faces)
print(len(faces))
print(faces.shape)

for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    print(x,"--",y,"--",w,"--",h)
    break
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
cv.imshow('img',img)
img1=img[faces[1,1]:faces[1,1]+105,faces[1,1]:faces[1,1]+105]
plt.imshow(img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

vr='C:\\Users\\th\\Anaconda3\\envs\\env1\\Library\\etc\\haarcascades\\'
# face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalf
face_cascade = cv2.CascadeClassifier(vr+'haarcascade_frontalface_default.xml')
# eye_cascade = cv.CascadeClassifier(vr+'haarcascade_eye.xml')
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,0,0), 10)
    return face_img

cap = cv2.VideoCapture(0) 
while True: 
    
    ret, frame = cap.read(0) 
     
    frame = detect_face(frame)
 
    cv2.imshow('Video Face Detection', frame) 
 
    c = cv2.waitKey(1) 
    if c == 27: 
        break 
        
cap.release() 
cv2.destroyAllWindows()
```
